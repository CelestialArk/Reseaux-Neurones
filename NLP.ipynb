```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
#Importer la librarie de lecture de PDF
import PyPDF2

#Ouvrir Le PDF
with open('/content/sample_data/Khaldi Achraf CV (fr).pdf', 'rb') as cv:
    reader = PyPDF2.PdfReader(cv)

    #Initier un string pour le text extrait du cv
    text_cv = ""

    for page in reader.pages:
        text_cv += page.extract_text()
```

```python
#Afficher le resultat de l'importation du pdf
print(text_cv)
```

```python
#Initiliser le NTLK pour tokenizer notre cv
nltk.download("punkt")

cv_tokens = word_tokenize(text_cv, language="french")
```

```python
#Afficher le resultat de la tokenization de notre cv
print(cv_tokens)
```

```python
#On remarque qu'il y a des ponctuations dans notre tokenization il faut les filtrer
cv_tokens_filtrer = [token for token in cv_tokens if token.isalnum()]

print(cv_tokens_filtrer)
```

```python
#Declarer la variable offre_stage qui contient le texte a comparer
offre_stage = "Le candidat idéal doit : Avoir 3 ans d'expérience minimum en HTML5, CSS3 et Javascript (Jquery) Avoir une expérience avec React (Next.js, Astro.js), Redux Avoir des connaissances en Tailwind CSS Avoir des connaissances en normes W3C, en CMS Avoir des connaissances des principes de SEO Expérience avec Git Ainsi que des compétences en curiosité, force de proposition et travail d'équipe."

#Tokenizer le text
offre_tokens = word_tokenize(offre_stage, language="french")
```

```python
#Afficher le resultat de la tokenization de L'offre
print(offre_tokens)
```

```python
#On constate le meme probleme donc on filtre de la meme maniere
offre_tokens_filtrer = [token for token in offre_tokens if token.isalnum()]

print(offre_tokens_filtrer)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

# Convertir les listes de mots clés en chaînes de texte
mots_cles_offre_stage = ' '.join(offre_tokens_filtrer)
mots_cles_cv = ' '.join(cv_tokens_filtrer)

# Créez un vectoriseur de texte pour vectoriser le text XD
vectorizer = CountVectorizer().fit_transform([mots_cles_offre_stage, mots_cles_cv])

# Et puis On calcule la similarite
similarite = cosine_similarity(vectorizer)
```

```python
#On affiche donc notre resultat
print(similarite)
```

```python
#On peut aller plus loin en affichant les mots qui se repete dans notre cv et l'offre de stage
mots_cles_communs = set(offre_tokens_filtrer) & set(cv_tokens_filtrer)
print(mots_cles_communs )
```
