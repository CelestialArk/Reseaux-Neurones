```python
#Importation du fichier text
text_file = open("/content/sample_data/text", "r" )

#Lecture du fichier
text = text_file.read()

print(text)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download("wordnet")
nltk.download("punkt")
nltk.download("stopwords")

tokenize_result = word_tokenize(text)

print(tokenize_result)
```

```python
#On elimine les elements intrus
tokens_filtrer = [token for token in tokenize_result if token.isalnum()]

print(tokens_filtrer)
```

```python
#On applique la lemantization sur nos donnees
lemmantizer = WordNetLemmatizer()

lemantized_Verb = [lemmantizer.lemmatize(word, pos="v") for word in tokens_filtrer]

lemantized_Noun = [lemmantizer.lemmatize(word, pos="n") for word in lemantized_Verb]

lemantized_Adj = [lemmantizer.lemmatize(word, pos="a") for word in lemantized_Noun]

lemantized = lemantized_Adj

print(lemantized)
```
